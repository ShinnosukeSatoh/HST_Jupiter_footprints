```python
def HUE23(S3wlong, hem):

    if hem=='N':
        delta = 5.78+3.29*math.cos(0.99*math.radians(S3wlong)) \
                    + 1.41*math.sin(0.99*math.radians(S3wlong))

    else:
        delta = 7.06-3.37*math.cos(math.radians(S3wlong)) \
                    + 0.15*math.sin(0.99*math.radians(S3wlong))
        
    return delta
```
